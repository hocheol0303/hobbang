### 과제 개요

### 과제 출제 목적 및 배경
시각화는 데이터 분석 이후 결과를 나타내기 위해 사용되기도 하지만, 데이터 분석 전 데이터의 특성을 파악하기 위해서도 사용됩니다. 데이터의 특성을 파악하는 탐색적 데이터 분석은 데이터 분석의 방향성을 결정하는 데에 있어 필수적이면서도 중요한 과정이며, 데이터에 따라 다양한 그래프 시각화 능력이 요구됩니다. 본 과제에서는 데이터를 불러온 후 다양한 그래프를 활용해서 직접 데이터의 구성을 시각화해보며 탐색적 데이터 분석을 수행해보는 것을 목표로 합니다.
