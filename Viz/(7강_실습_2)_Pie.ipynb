# (7강-실습-2) 고급시각화 (Pie Chart)

## 0. 들어가기 앞서

- **실습 개요**: Pie chart와 Bar chart에 대해 알아본다.
- **실습 진행 목적 및 배경**:
  - Pie chart와 Bar chart를 직접 구현해보고, 특성을 비교해볼 수 있음
  - Pie chart의 다양한 변형을 구현해볼 수 있음
- **실습 수행으로 얻어갈 수 있는 역량**:
  - 다양한 Pie chart와 Bar chart에 대한 이해
- **데이터셋 개요 및 저작권 정보**:
  - None
- **실습 핵심 내용**
  - matplotlib를 통한 Pie chart와 Bar chart에 대한 이해
  - matplotlib를 통한 Pie chart의 다양한 변형



```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## 1. Pie Chart & Bar Chart

### 1-1. Basic Pie Chart

- `pie()`

```python
labels = ['A', 'B', 'C', 'D']
data = np.array([60, 90, 45, 165]) # total 360
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,12))
ax[0].pie(data
       ,labels=labels
      )
ax[1].pie(data[::-1], labels=labels[::-1])
plt.show()
```

### 1-2. Pie Chart vs Bar Chart

같은 데이터로 Pie chart와 Bar Chart를 비교하며 장단점을 비교하면 다음과 같습니다.

- 장점 : 비율 정보에 대한 정보를 제공할 수 있다.
- 단점 : 구체적인 양의 비교가 어렵다.

```python
# pie chart는 양의 비교는 어려움. 막대그래프가 더 쉬움
fig, axes = plt.subplots(1, 2, figsize=(12, 7))

axes[0].pie(data, labels=labels)
axes[1].bar(labels, data)

plt.show()
```

차이가 큰 데이터에 대해서는 비교를 할 수 있지만, 다음과 같이 비슷한 값들에 대해서는 비교가 어렵습니다.

```python
np.random.seed(97)

data = np.array([16, 18, 20, 22, 24])
labels = list('ABCDE')
color = plt.colormaps.get_cmap('tab10').colors[:5]
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i in range(3):
    axes[0][i].pie(data, labels=labels)
    axes[1][i].bar(labels, data, color=color)
    np.random.shuffle(data)

plt.show()
```

### 1-2. Pie Chart Custom

- `startangle`

```python
labels = ['A', 'B', 'C', 'D']
data = np.array([60, 90, 45, 165]) # total 360
```

```python
# startangle: 시작점 변경
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax.pie(data, labels=labels,
        startangle=90
      )
plt.show()
```

- `explode`

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
explode = [0, 0, 0.2, 0]

ax.pie(data, labels=labels, explode=explode, startangle=90)
plt.show()

```

- `shadow`


```python
# 시각화에서 그림자는 좋은 방법 아님
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
explode = [0, 0, 0.2, 0]

ax.pie(data, labels=labels, explode=explode, startangle=90,
      shadow=True)
plt.show()

```

- `autopct`

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
explode = [0, 0, 0.2, 0]

ax.pie(data, labels=labels, explode=explode, startangle=90,
      shadow=True, autopct='%1.1f%%')
plt.show()

```

- `labeldistance`
- `rotatelabels`


```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
explode = [0, 0, 0.2, 0]

ax.pie(data, labels=labels, explode=explode, startangle=90,
      shadow=True, autopct='%1.1f%%', labeldistance=1.15
#        , rotatelabels=90
      )
plt.show()

```

- `counterclock`

```python
# 시계방향
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
explode = [0, 0, 0.2, 0]

ax.pie(data, labels=labels, explode=explode, startangle=90, counterclock=False)
plt.show()

```

- `radius`

```python
# 크기 조절
fig, axes = plt.subplots(1, 3, figsize=(12, 7))
explode = [0, 0, 0.2, 0]

for size, ax in zip([1, 0.8, 0.5], axes):
    ax.pie(data, labels=labels, explode=explode, startangle=90, counterclock=False,
           radius=size
          )
plt.show()

```

## 2. Pie Chart 변형

### 2-1. Donut Chart

중간에 원을 그리는 방식으로 그려집니다.

```python
# 도넛차트
fig, ax = plt.subplots(1, 1, figsize=(7, 7))


ax.pie(data, labels=labels, startangle=90,
      shadow=True, autopct='%1.1f%%')              # 비율 표시

# 좌표 0, 0, r=0.7, facecolor='white'
centre_circle = plt.Circle((0,0),0.70,fc='white')  # [0,0] 차트에 70% 크기의 흰색 원을 그릴거
ax.add_artist(centre_circle)

plt.show()
```

- `pctdistance`
- `textprops`

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))


ax.pie(data, labels=labels, startangle=90,
      shadow=True, autopct='%1.1f%%', pctdistance=[0.85,0.2][0], textprops={'color':["w",'r'][0]})       # pctdistance: 중앙과의 거리 -> 밖으로 밀어내기

# 좌표 0, 0, r=0.7, facecolor='white'
centre_circle = plt.Circle((0,0),0.70,fc='white')
ax.add_artist(centre_circle)

plt.show()
```
